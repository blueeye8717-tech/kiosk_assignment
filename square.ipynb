```python
class Square:
    def __init__(self, shape_type:int):
        """1=직사각형, 2=평행사변형, 3=사다리꼴"""
        self.square = shape_type
        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')
        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self, width:int, vertical:int) -> str:
        area = width * vertical
        return f'직사각형의 넓이는 : {area}'

    def par(self, base:int, height:int) -> str:
        area = base * height
        return f'평행사변형의 넓이는 : {area}'

    def trape(self, top:int, bottom:int, height:int) -> str:
        area = (top + bottom) * height / 2
        return f'사다리꼴의 넓이는 : {area}'
```
