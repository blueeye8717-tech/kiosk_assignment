```python
class Kiosk:    
    def __init__(self):
        # 메뉴 및 가격 리스트
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.order_list = []     # 주문한 메뉴 이름 저장
        self.order_price = []    # 주문한 메뉴 가격 저장
        self.price_sum = 0       # 총 금액

    # 메뉴 출력 메서드
    def menu_print(self):
        print("\n메뉴판")
        print("-------------------")
        for i in range(len(self.menu)):
            print(f"{i+1}. {self.menu[i]} : {self.price[i]}원")
        print("-------------------")

    # 주문 메서드
    def menu_select(self):
        print()
        n = 0
        while n < 1 or len(self.menu) < n:
            n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력

            # 메뉴판에 있는 음료 번호일 때
            if 1 <= n <= len(self.menu):
                self.price_sum += self.price[n-1]  # 선택 음료의 가격 누적
                self.order_list.append(self.menu[n-1])   # 메뉴 이름 추가
                self.order_price.append(self.price[n-1]) # 메뉴 가격 추가
                print(self.menu[n-1], " : ", self.price[n-1], "원")
                print(f"현재까지 합계 : {self.price_sum}원")

                # 추가 주문 여부 확인
                add = input("다른 음료를 추가 주문하시겠습니까? (y/n) : ")
                if add.lower() == 'y':
                    self.menu_print()
                    continue
                else:
                    break
            
            # 메뉴판에 없는 번호일 때
            else:  
                print("없는 메뉴입니다. 다시 주문해 주세요.")
        return self.price_sum

    # 지불 메서드
    def pay(self):
        print("\n결제 단계로 넘어갑니다.")
        print(f"총 금액은 {self.price_sum}원입니다.")
        money = int(input("지불 금액을 입력하세요 : "))

        if money < self.price_sum:
            print(f"금액이 부족합니다. {self.price_sum - money}원이 모자랍니다.")
            return False
        else:
            change = money - self.price_sum
            print(f"결제가 완료되었습니다. 거스름돈은 {change}원입니다.")
            return True

    # 주문표(영수증) 작성 메서드
    def table(self):
        print("\n======= 주문표 =======")
        for i in range(len(self.order_list)):
            print(f"{i+1}. {self.order_list[i]} : {self.order_price[i]}원")
        print("-------------------------")
        print(f"총 결제 금액 : {self.price_sum}원")
        print("=========================\n")
```
